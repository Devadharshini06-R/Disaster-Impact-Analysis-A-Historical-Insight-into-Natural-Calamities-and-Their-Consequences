```python
def convert_damages(MONEY_LOST_USD_MILLION):
    endresult = []
    for i in MONEY_LOST_USD_MILLION:
        if pd.isna(i):
            endresult.append(None)
            continue
        parts = i.strip().replace('$', '').split()
        num = float(parts[0])
        unit = parts[1].lower()
        if unit == 'billion':
            result = num * 1_000_000_000
        elif unit == 'million':
            result = num * 1_000_000
        else:
            result = None
        endresult.append(result)
    return endresult
```

```python
damage_converted=convert_damages(df['MONEY_LOST_USD_MILLION'])
```
